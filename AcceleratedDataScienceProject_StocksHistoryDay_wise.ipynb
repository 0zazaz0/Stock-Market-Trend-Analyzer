```python
def pred_(comp_):

  df = pd.DataFrame(stocksDict[comp_]['Close'])
  df = df.dropna(axis = 0)
  # df = df.tail(30)._append(df,ignore_index=True)
  df = scaler.fit_transform(df)

  x_test = []
  y_test = []
  for i in range(30,df.shape[0]):
    x_test.append(df[i-30:i])
    y_test.append(df[i,0])
  x_test, y_test = np.array(x_test), np.array(y_test)

  y_preds = model.predict(x_test)
  ev = model.evaluate(x_test,y_test)
  loss = ev[0]
  rmse = ev[1]
  mape = ev[2]
  r2 = ev[3]

  factor = scaler.scale_
  factor = 1/factor
  y_preds = y_preds*factor
  y_test = y_test*factor

  plt.figure(figsize = (12,6))
  plt.plot(y_preds, 'g', label = 'predictions')
  plt.plot(y_test, 'r',label = 'original')
  plt.xlabel(f'day waise from JANUARY-2010 to DECEMBER-2023')
  # plt.ylabel('stock prices (RUPEES)')
  # plt.legend()
  plt.title(f'Stock prices of {comp_}' )
  plt.grid(axis='both', alpha=.3)
  plt.legend()
  print(f'LOSS: {loss}\nRMSE: {rmse}\nMAPE: {mape}\nR2: {r2}')
  return ev
```
